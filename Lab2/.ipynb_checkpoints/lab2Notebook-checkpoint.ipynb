```python
# Trochę magii
%matplotlib inline
from matplotlib import pyplot as plt
from sympy.interactive import printing
printing.init_printing(use_latex=True)
import numpy as np
from __future__ import division
import sympy as sym
from sympy import *

# Definicje zmiennych
x, y, z = symbols("x y z")
k = Symbol("k", integer=True)
f = Function('f')
f
```

```python
xRange = np.arange(-5, 5, 0.1)
func = -(x ** 3) + 3 * x ** 2 + 10 * x - 24
res = [func.subs(x, i).evalf() for i in xRange]
plt.plot(xRange, res)
plt.show()
exp = solveset(func, x)
exp
```

```python
fun1 = x**2 + 3*y - 10
fun2 = 4*x - y**2 + 2
sol = sym.solve([fun1, fun2])
sol
```

```python
for s in sol:
    for k, v in s.items():
        print(k, v.evalf())
```

```python
fun5 = sin(log(x, 2)) * cos(x**2)/x
Derivative(fun5)
```
